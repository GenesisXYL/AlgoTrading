# Merge TAQ

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# load our helper functios
from simtools import *

matplotlib.rcParams[ 'figure.figsize' ] = ( 16, 9 )
```

```python
# set up some variables
trade_file = 'TSLA_T.csv'
quote_file = 'TSLA_Q.csv'
```

```python
# call load trades
trades = loadtradefile(trade_file)
```

```python
trades.tail()
```

```python
# call load quotes
quotes = loadquotefile(quote_file)
```

```python
quotes.tail()
```

```python
# 09:30 - 16:10
def select_trade_time(tick_data):
    tick_data=tick_data[tick_data.index.hour<16]
    tick_data=tick_data[(tick_data.index.hour*60+tick_data.index.minute)>=(9*60+30)]
    return tick_data

trades=select_trade_time(trades)
quotes=select_trade_time(quotes)

# merge the files
taq = makeTAQfile(trades, quotes)
```

```python
taq.tail(20)
```
